```python
import pandas as pd

# Load the dataset
df = pd.read_csv("ecommerce_return_augmented.csv")

# === HANDLE MISSING VALUES ===
df['Return_Date'] = df['Return_Date'].fillna('Not Returned')
df['Return_Reason'] = df['Return_Reason'].fillna('No Return')
df['Days_to_Return'] = df['Days_to_Return'].fillna(0)
df['Region'] = df['Region'].fillna(df['Region'].mode()[0])

# === CONVERT DATE COLUMNS ===
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Return_Date'] = pd.to_datetime(df['Return_Date'], errors='coerce')

# === CONVERT BOOLEAN COLUMNS TO INTEGER ===
df['Product_Returnable'] = df['Product_Returnable'].astype(int)
df['High_Return_Risk_Flag'] = df['High_Return_Risk_Flag'].astype(int)

# === EXTRACT DATE FEATURES ===
df['Order_Weekday'] = df['Order_Date'].dt.day_name()  # Gives 'Monday', 'Tuesday', etc.
df['Order_Month'] = df['Order_Date'].dt.month

# Preview processed data
df.head()
```

```python
df.to_csv("ecommerce_return_cleaned.csv", index=False)
```
